Here's a summary of the key takeaways from the comparison:

**Feedforward NN** is stateless and the simplest of the four. It's fast and parallelizable, but it has no concept of time or sequence — each input is treated independently. Best for tabular data and fixed-size inputs.

**RNN** introduces recurrence — a hidden state carries information from one step to the next. This allows it to model short-term dependencies in sequences, but it famously suffers from the vanishing gradient problem, meaning it struggles to learn from events far back in a sequence.

**LSTM** was specifically designed to fix the RNN's vanishing gradient problem. Its three gates (forget, input, output) and separate cell state let it selectively remember or discard information over long sequences. It's much more capable for NLP, speech, and time-series tasks — at the cost of more parameters and slower training.

**Bidirectional LSTM** runs two LSTMs simultaneously — one forward and one backward — and concatenates their outputs. This gives every timestep access to both past and future context, which is powerful for tasks like named entity recognition or sentiment analysis where the full sentence matters. The tradeoff is that it requires the complete sequence upfront (no real-time streaming) and doubles the parameter count and compute cost.
